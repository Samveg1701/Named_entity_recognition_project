```python
import nltk
import pandas as pd

# Function to perform NER and POS tagging for a given text
def perform_ner_pos(text):
    # Split the text into words using spaces
    words = text.split()
    
    # Perform POS tagging
    pos_tags = nltk.pos_tag(words)
    
    # Perform named entity recognition
    ner_tags = nltk.ne_chunk(pos_tags)
    
    # Convert the named entity chunk tree to IOB tags
    iob_tags = nltk.tree2conlltags(ner_tags)
    tag_map = {'B-ORGANIZATION': 'B-org', 'I-ORGANIZATION': 'I-org',
               'B-PERSON': 'B-per', 'I-PERSON': 'I-per',
               'B-LOCATION': 'B-geo', 'I-LOCATION': 'I-geo',
               'B-GPE': 'B-gpe', 'I-GPE': 'I-gpe',
               'B-FACILITY': 'B-fac', 'I-FACILITY': 'I-fac',
               'B-GSP': 'B-gsp', 'I-GSP': 'I-gsp',
               'B-DATE': 'B-date', 'I-DATE': 'I-date',
               'B-TIME': 'B-tim', 'I-TIME': 'I-tim',
               'B-MONEY': 'B-money', 'I-MONEY': 'I-money',
               'B-NUMBER': 'B-num', 'I-NUMBER': 'I-num',
               'O': 'O'}
    
    iob_tags_shortened = [(word, pos, tag_map.get(tag, tag)) for word, pos, tag in iob_tags]
    
    # Extract only the NER tags
    ner_tags = [tag for (_, _, tag) in iob_tags_shortened]
    
    # Extract only the POS tags
    pos_tags = [pos_tag for (_, pos_tag) in pos_tags]
    
    return ner_tags, pos_tags

# Load the dataset
df = pd.read_csv('NLP_task_2.csv').head(100)

# Create new columns to store NER and POS tags
df['NER_Tags'] = ''
df['POS_Tags'] = ''

# Iterate over each row in the dataset
for index, row in df.iterrows():
    # Get the text from the current row
    text = row['text']
    
    # Perform NER and POS tagging for the text
    ner_tags, pos_tags = perform_ner_pos(text)
    
    # Store the NER and POS tags in the corresponding row of the dataset
    df.at[index, 'NER_Tags'] = str(ner_tags)
    df.at[index, 'POS_Tags'] = str(pos_tags)

# Save the updated dataset
df.to_csv('updated_dataset_with_ner_pos.csv_q2_nltk', index=False)

print("NER and POS tags stored separately for each row in updated_dataset_with_ner_pos.csv")
```

```python
import jsonlines

# Open the JSONL file for reading
with jsonlines.open('POS_tagging.jsonl') as reader:
    # Initialize an empty list to store the third values
    POS_tags_q2 = []
    
    # Iterate over each line in the JSONL file
    for line in reader:
        # Access the 'label' key which contains a list of lists
        labels = line['label']
        
        # Iterate over each sublist
        for sublist in labels:
            # Check if the sublist has at least three elements
            if len(sublist) >= 3:
                # Append the third value to the list of third values
                POS_tags_q2.append(sublist[2])
    
    # Find the index of the first occurrence of 'NN' tag
    index_of_first_nn = POS_tags_q2.index('NN')
    
    # Split the list before the first 'NN' tag
    NER_tags_q2 = POS_tags_q2[:index_of_first_nn]
    POS_tags_q2 = POS_tags_q2[index_of_first_nn:]
    
# Now split_list contains the third values before the first 'NN' tag
print(NER_tags_q2)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Load the dataset
df = pd.read_csv('updated_dataset_with_ner_pos.csv')

# Convert string representations of lists to actual lists
df['Tag'] = df['Tag'].apply(eval)
df['NER_Tags'] = df['NER_Tags'].apply(eval)

# Flatten the lists of tags
actual_tags = [tag for sublist in df['Tag'] for tag in sublist]
predicted_tags = [tag for sublist in df['NER_Tags'] for tag in sublist]

# Create the confusion matrix
conf_matrix = confusion_matrix(actual_tags, predicted_tags)

# Define labels for the confusion matrix
labels = sorted(set(actual_tags + predicted_tags))

# Create a DataFrame from the confusion matrix and labels
conf_df = pd.DataFrame(conf_matrix, index=labels, columns=labels)

# Set up the plot
plt.figure(figsize=(12, 10))
sns.set(font_scale=1.2)

# Plot the heatmap with a custom color map
heatmap = sns.heatmap(conf_df, annot=True, fmt='d', cmap="Greys", cbar=False, linewidths=.5)

# Set labels and title
heatmap.set_xlabel('Predicted Labels', fontsize=16)
heatmap.set_ylabel('Actual Labels', fontsize=16)
heatmap.set_title('Confusion Matrix', fontsize=18)

# Rotate the tick labels for better readability
plt.xticks(rotation=45)
plt.yticks(rotation=0)

# Show the plot
plt.show()
```

```python
import pandas as pd
from sklearn.metrics import confusion_matrix

# Load the dataset
df = pd.read_csv('updated_dataset_with_ner_pos.csv')

# Convert string representations of lists to actual lists
df['Tag'] = df['Tag'].apply(eval)
df['NER_Tags'] = df['NER_Tags'].apply(eval)

# Flatten the lists of tags
actual_tags = [tag for sublist in df['Tag'] for tag in sublist]
predicted_tags = [tag for sublist in df['NER_Tags'] for tag in sublist]

# Calculate confusion matrix
conf_matrix = confusion_matrix(actual_tags, predicted_tags, labels=sorted(set(actual_tags)))

# Initialize dictionaries to store metrics for each class
class_metrics = {'TP': {}, 'FP': {}, 'TN': {}, 'FN': {}}

# Calculate TP, FP, TN, FN for each class
for i, label in enumerate(sorted(set(actual_tags))):
    TP = conf_matrix[i, i]
    FP = conf_matrix[:, i].sum() - TP
    FN = conf_matrix[i, :].sum() - TP
    TN = conf_matrix.sum() - (TP + FP + FN)
    class_metrics['TP'][label] = TP
    class_metrics['FP'][label] = FP
    class_metrics['TN'][label] = TN
    class_metrics['FN'][label] = FN

# Create DataFrames for TP, FP, TN, FN for each class
metrics_dfs = {metric: pd.Series(values) for metric, values in class_metrics.items()}
metrics_df = pd.DataFrame(metrics_dfs)

print("Matrix of True Positives, False Positives, True Negatives, and False Negatives:")
print(metrics_df)
```
